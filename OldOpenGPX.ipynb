```python
import pandas as pd
from b2.gsas import GSASModel
```

```python
AlNi_model = GSASModel("Al-Ni-dual-phase.gpx")
AlNi_phases = AlNi_model.phases
AlNi_phases
```

```python
AlNi_model.derivative(["0:0:Size;i"])
```

```python
# plot default gpx file
xrd_simul = AlNi_model.forward()
import matplotlib.pyplot as plt
plt.plot(xrd_simul)
plt.xlabel('2theta')
plt.ylabel('Intensity')
plt.title('Default AlNi Model')
```

```python
# change microstrain or grain size
xrd_simul1 = AlNi_model.forward(mustrain=50000, grainsize = 0.07)
plt.plot(xrd_simul1)
```
